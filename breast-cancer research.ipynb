```python
from sklearn.model_selection import train_test_split
train, test=train_test_split(df, test_size=0.2, random_state=42)
print("training data:", len(train))
print("testing data:", len(test))
```

```python
train_label=train["diagnosis"].copy()
train=train.drop(columns=["diagnosis"], axis=1)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
my_pipe=Pipeline([
    ('std_scaler',StandardScaler())
])
```

```python
train_1=my_pipe.fit_transform(train)
train_1
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn import svm
from sklearn.cluster import KMeans
#model=DecisionTreeRegressor()
#model=RandomForestRegressor()
#model=KNeighborsClassifier(n_neighbors=4)
model=LogisticRegression(solver='liblinear')
#model=svm.SVC(kernel="rbf")
#model=KMeans(init="k-means++", n_clusters=4, n_init=12)
model.fit(train, train_label)
```

```python
from sklearn.metrics import mean_squared_error
yhat=model.predict(train)
mse=mean_squared_error(train_label, yhat)
rmse=np.sqrt(mse)
```

```python
mse 
```

```python
from sklearn.model_selection import cross_val_score
scores=cross_val_score(model,train, train_label,scoring="neg_mean_squared_error", cv=10)
rmse_scores=np.sqrt(-scores)
```

```python
rmse_scores
```

```python
def print_scores(scores):
    print("Scores:", scores)
    print("Mean:", scores.mean())
    print("Standard Deviation:", scores.std())
```

```python
print_scores(rmse_scores)
```
